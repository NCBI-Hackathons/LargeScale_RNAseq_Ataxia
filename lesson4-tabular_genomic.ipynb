```
import pandas as pd
url='https://github.com/NCBI-Hackathons/MassiveSeq/blob/master/atax_expr_fastai.csv?raw=true'
df2=pd.read_csv(url)
```

```
catnames=df2.columns[2:53].tolist()
contnames=df2.columns[54:].tolist()
```

```
test=TabularList.from_df(df2.iloc[70:].copy(),path=path,cat_names=catnames,cont_names=contnames)
```

```
data = (TabularList.from_df(df2, path=path, cat_names=catnames, cont_names=contnames,procs=procs)
                           .split_by_idx(list(range(70,82)))
                           .label_from_df(cols=dep_var)
                           .add_test(test)
                           .databunch())
```
